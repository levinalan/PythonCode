```python
def show_density(data):
    fig = plt.figure(figsize=(10,4))
    data.plot.density()

    plt.axvline(x=min_val, color = 'gray')
    plt.axvline(x=max_val, color = 'gray')
    plt.axvline(x=mean_val, color = 'cyan')
    plt.axvline(x=med_val, color = 'red')
    plt.axvline(x=mode_val, color = 'yellow')

    plt.title('Data Density')
    plt.show()

show_density(df_students_nu.Grade)
```

```python
import math
mean = df_students_nu.Study_Hours.mean()
variance = 0
for hours in df_students_nu.Study_Hours:
    variance = variance + (mean - hours)**2

variance = variance/df_students_nu.shape[0]

print(variance)

std = math.sqrt(variance)

print(std)

for col_name in ['Grade', 'Study_Hours']:
    col = df_students_nu[col_name]
    rng = col.max() - col.min()
    var = col.var()
    std = col.std()
    print(f'\n{col_name}:\n - Range: {rng}\n - variance: {var}\n - Std Dev: {std}')
    print(f'68 % of gradew will fall between {df_students_nu.Grade.mean() - df_students_nu.Grade.std()} and {df_students_nu.Grade.mean() + df_students_nu.Grade.std()}')

df_students_nu.describe()
```
